```python
def binary_to_int(bin_str):
    return int(bin_str, 2)

def int_to_binary(num, length):
    return bin(num)[2:].zfill(length)

def floating_point_add_sub(n, m, S1, S2, E1, E2, flag):
    # Convert inputs to integers
    mant1 = 1 + binary_to_int(S1) / (2 ** n)
    mant2 = 1 + binary_to_int(S2) / (2 ** n)
    exp1 = binary_to_int(E1)
    exp2 = binary_to_int(E2)

    # Apply sign change if subtraction
    if flag == 0:  # Subtraction
        mant2 = -mant2

    # Align exponents
    if exp1 > exp2:
        mant2 /= (2 ** (exp1 - exp2))
        exponent = exp1
    elif exp2 > exp1:
        mant1 /= (2 ** (exp2 - exp1))
        exponent = exp2
    else:
        exponent = exp1

    # Perform addition
    result_mantissa = mant1 + mant2

    # Normalize result
    if result_mantissa == 0:
        return "0" * (n + m)  # Return binary zero result

    while result_mantissa >= 2:
        result_mantissa /= 2
        exponent += 1
    while result_mantissa < 1:
        result_mantissa *= 2
        exponent -= 1

    # Convert mantissa to binary
    result_mantissa_bin = int_to_binary(int((result_mantissa - 1) * (2 ** n)), n)
    result_exponent_bin = int_to_binary(exponent, m)

    return f"1.{result_mantissa_bin} x 2^{result_exponent_bin}"


n = int(input("Enter the number of bits for the significand (n): "))
m = int(input("Enter the number of bits for the exponent (m): "))
S1 = input(f"Enter the {n}-bit string for significand of A: ")
S2 = input(f"Enter the {n}-bit string for significand of B: ")
E1 = input(f"Enter the {m}-bit string for exponent of A: ")
E2 = input(f"Enter the {m}-bit string for exponent of B: ")
flag = int(input("Enter 1 for addition or 0 for subtraction: "))
result = floating_point_add_sub(n, m, S1, S2, E1, E2, flag)
print("Result in binary format:", result)
```

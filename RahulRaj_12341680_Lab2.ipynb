2. Problem: Write a Python program that finds the GCD (Greatest Common Divisor) and LCM (Least Common Multiple) of all numbers in a given range [a, b] where a <= b

```python
import math

def gcd(a, b):
    if(a==0): return b
    return gcd(b%a,a)

def find_gcd_lcm_of_range(start, end):
    # current_gcd=gcd(start,end)
    # current_lcm=(start*end)//(current_gcd)
    # return current_gcd, current_lcm
    gcd_result=[]
    lcm_result=[]
    for i in range(start,end+1):
      for j in range(i+1,end+1):
        hcf=gcd(i,j)
        gcd_result.append({f"{i} and {j} ",hcf})
        lcm=(i*j)//hcf
        lcm_result.append({f"{i} and {j} ",lcm})

    return gcd_result,lcm_result


start = int(input("Enter the start number: "))
end = int(input("Enter the end number: "))
gcd_result, lcm_result = find_gcd_lcm_of_range(start, end)
print(f"The GCD of numbers between {start} and {end} is {gcd_result}")
print(f"The LCM of numbers between {start} and {end} is {lcm_result}")
```

3. Problem: Write a Python program to solve the quadratic equation of the form ax^2 + bx + c = 0 for given a, b, and c. The program should return the two roots (if they exist) or a message if the equation has no real roots.

```python
import math

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4 * a * c
    if discriminant > 0:
       #Implement your code here....
       root1=(-b + math.sqrt(discriminant))/(2*a)
       root2=(-b-math.sqrt(discriminant))/(2*a)
       return {root1,root2}

    elif discriminant == 0:
      roots= -b/(2*a)
      return {roots,roots}

    else:
        return "No real roots"


a, b, c = 1, -3, 2
result = solve_quadratic(a, b, c)
print(f"The roots are: {result}")
```

4. Problem: Write a Python program to generate all Pythagorean triplets (a, b, c) where a^2 + b^2 = c^2 for a <= b <= c <= n.


```python
def pythagorean_triplets(n):
    triplets = []
    for a in range(1, n + 1):
        for b in range(a, n + 1):
            for c in range(b, n + 1):
                if a**2 + b**2 == c**2:
                    triplets.append((a, b, c))
    return triplets


n = 20
triplets = pythagorean_triplets(n)
print(f"Pythagorean triplets up to {n}: {triplets}")
```

6. Problem: Create a program that simulates a simple bank account system. The program should allow a user to perform the following operations:
a. Check balance
b. Deposit money
c. Withdraw money
d. Exit the program

```python
def bank_account_system():
    balance = 1000  # initial balance
    while True:
        print("\n--- Bank Account System ---")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
           #Implement your code here....
           print(f"Your Current Balance is :{balance}\n")



        elif choice == 2:
           #Implement your code here....
           deposite=int(input("Enter the money to deposite :"))
           balance +=deposite
           print("Money Deposited !\n")



        elif choice == 3:
            #Implement your code here....
            withdraw=int(input("Enter the money to withdraw :"))

            if(balance-withdraw < 0) :
              print("Less balance !!\n")
            else :
              balance -=withdraw
              print("Money Withdrawn !\n")



        elif choice == 4:
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

bank_account_system()
```

1. Problem: Find Longest Substring Without Repeating Characters


```python
def longest_substring(s):
    #Implement your code here....
    n = len(s)
    charSet = set()
    maxLen = 0
    l = 0

    for r in range(n):
        while s[r] in charSet:
            charSet.remove(s[l])
            l += 1
        charSet.add(s[r])
        maxLen = max(maxLen, r - l + 1)

    return maxLen


s = input("Enter a string: ")
print(f"The length of the longest substring without repeating characters is {longest_substring(s)}")
```

1. Calculate the mean, median, mode, standard deviation, and variance for the dataset:
[5.0, 10.0, 15.0, 20.0, 25.0, 30.0, 35.0]

```python
import statistics as stats
dataset=[5,10,15,20,25,30.35]
mean=stats.mean(dataset)
median=stats.median(dataset)
mode=stats.mode(dataset)
std_dev=stats.stdev(dataset)
variance=stats.variance(dataset)
print(f"Mean : {mean}")
print(f"Median : {median}")
print(f"Mode : {mode}")
print(f"Standard Deviation : {std_dev}")
print(f"Variance : {variance}")
```

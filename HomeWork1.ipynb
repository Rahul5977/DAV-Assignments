Question 10 [10] Computer Experiment: Find the average A1000000 of a million random 0-1 samples! What is the value of the standardized variable:X=(An-1/2)/(2(rootN)

```python
import numpy as np
arr=np.random.randint(0,2,size=1000000)
# print(arr)
Mean=np.mean(arr)
print(Mean)
X=(Mean-0.5)/(2*np.sqrt(1000000))
print(X)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

penguins = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')

# Basic EDA
# print(penguins.describe())
print(penguins)
penguins = penguins.dropna()
print(f"Head of the data\n:{penguins.head()}")
print(f"Tail of the data\n:{penguins.tail()}")
print(f"Shape :{penguins.shape}")
print(f"Discriptive Statistics{penguins.describe()}")

# Heatmap
# Selecting only numerical features for correlation calculation
numFeature = penguins.select_dtypes(include=['number'])
# print(numFeature)

plt.figure(figsize=(8, 6))
sns.heatmap(numFeature.corr(), annot=True, cmap='coolwarm') # Use numerical_features of penguins
plt.title('Correlation Heatmap of Penguin Features')
plt.show()

# Scatterplots
species_colors = {'Adelie': 'orange', 'Chinstrap': 'green', 'Gentoo': 'blue'}

plt.figure(figsize=(15,5))
plt.subplot(3,2,1)
sns.scatterplot(x='bill_length_mm', y='flipper_length_mm', hue='species', data=penguins, palette=species_colors)
plt.title('Bill Length vs. Flipper Length')

plt.subplot(3, 2, 2)
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=penguins, palette=species_colors)
plt.title('Bill Length vs. Bill Depth')

plt.subplot(3, 2, 3)
sns.scatterplot(x='body_mass_g', y='bill_length_mm', hue='species', data=penguins, palette=species_colors)
plt.title('Body Mass vs bill_lenght')

plt.subplot(3, 2, 4)
sns.scatterplot(x='body_mass_g', y='island', hue='species', data=penguins, palette=species_colors)
plt.title('Body Mass in diffirent islands')

plt.subplot(3, 2, 5)
sns.scatterplot(x='bill_length_mm', y='island', hue='species', data=penguins, palette=species_colors)
plt.title('Body Mass in diffirent islands')

plt.tight_layout()
plt.show()
```

```python
def add_sub(A, B, n, subtract):
    if subtract:
        # 2's comp of B for subtraction
        B = bin((~int(B, 2) + 1) & ((1 << n) - 1))[2:].zfill(n)

    # Add A and B
    res = bin(int(A, 2) + int(B, 2) & ((1 << n) - 1))[2:].zfill(n)
    return res

def div(dividend, divisor, n):
    A = '0'.zfill(n)
    M = divisor
    Q = dividend
    Count = n

    while Count > 0:
        A = (A[1:] + Q[0]).zfill(n) # Shiftleft A, add MSB of Q
        Q = (Q[1:] + '0').zfill(n)   # Shiftleft Q

        A = add_sub(A, M, n, True)

        if A[0] == '1':  # Check if MSB of A is 1 (negative)
            Q = Q[:-1] + '0'  # Set Q_0 to 0
            A = add_sub(A, M, n, False)
        else:
            Q = Q[:-1] + '1'  # Set Q_0 to 1

        Count -= 1

        print(f"After iteration: A={A}, Q={Q}")

    print(f"Quotient in Q: {Q}")
    print(f"Remainder in A: {A}")

dividend = input("Enter dividend in binary 2's complement: ")
divisor = input("Enter divisor in binary 2's complement: ")
n = int(input("Enter number of bits: "))

div(dividend, divisor, n)
```

Create an empty graph G.

Add the following nodes: A, B, C, D

Add the following edges: A-B, B-C, C-D, D-A, B-D

```python
import networkx as nx
# An empty undirected graph
G = nx.Graph()

# Adding the following nodes A,B,C,D
G.add_nodes_from(['A','B','C','D'])

# Adding the following edges A-B, B-C, C-D, D-A, B-D
G.add_edges_from([('A','B'),('B','C'),('C','D'),('D','A'),('B','D')])
```

Visualize the Graph

```python
nx.draw(G,with_labels=True)
```

Add an attribute color to node A with value 'red'.

Add a weight attribute to edge (B, D) with value 5.


```python
# adding an attribute color to node A with value red
nx.set_node_attributes(G, 'red', 'A')
# add a weight attribute to edge (B,D) with value 5.
G.add_weighted_edges_from([('A','B',1),('B','C',1),('C','D',1),('D','A',1),('B','D',5)])

for (u, v, wt) in G.edges.data('weight'):
	print(f"({u}, {v}, {wt})")
nx.draw(G,with_labels=True)
```

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)  # Define layout
nx.draw(G, pos, with_labels=True)  # Draw nodes and edges
edge_labels = nx.get_edge_attributes(G, 'weight')  # Get edge weights
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)  # Draw edge labels

plt.show()  # Display the graph
```

How many nodes are there in the graph?

How many edges are there?

List all the neighbors of node B.

```python
# print number of nodes
print(f"Number of nodes :{G.number_of_nodes()}")
# print number of edges
print(f"Number of edges :{G.number_of_edges()}")
# print neighbors of node B
print(f"Neighbors of B :{list(G.neighbors('B'))}")
```

Find the shortest path between node A and node C (assuming unweighted edges).


```python
G = nx.Graph()

G.add_nodes_from(['A','B','C','D'])

G.add_edges_from([('A','B'),('B','C'),('C','D'),('D','A'),('B','D')])

# find shortest path
print(f"Shortest path between A and C :{nx.shortest_path(G,'B','C')}")
```

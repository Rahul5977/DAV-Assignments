```python
# (b) The team wants to explore how different response probabilities might affect campaign performance.
# Repeat part (a) by considering probabilities of conversion as 10%, 20%, ..., 90%. Analyze and
# comment on how the distribution changes as the probability of a successful response increases.
# Different probability values to analyze
n = 15  # Number of trials
p_values = np.arange(0.1, 1.0, 0.1)  # Different probabilities
x = np.arange(0, n+1)  # Possible number of successes

# Create separate plots for each probability
fig, axes = plt.subplots(len(p_values), 1, figsize=(8, 3 * len(p_values)), sharex=True)

for i, p in enumerate(p_values):
    pmf = binom.pmf(x, n, p)
    axes[i].stem(x, pmf)
    axes[i].set_title(f'PMF for p={p:.1f}')
    axes[i].set_ylabel('PMF')
    axes[i].grid(True)

axes[-1].set_xlabel('Number of Successful Responses')
plt.tight_layout()
plt.show()
```

```python
# Campaign sizes
n_values = [10, 20, 50, 200]
p = 0.05  # Conversion probability

# Generate plots for success rate distributions
plt.figure(figsize=(10, 6))
x_range = np.linspace(0, 1, 100)  # Success rate range
colors = ['b', 'g', 'r', 'm']

for n,color in zip(n_values,colors):
    x_vals = np.arange(0, n+1)  # Possible number of successes
    pmf_vals = binom.pmf(x_vals, n, p)  # Binomial PMF
    success_rates = x_vals / n  # Convert X to Y = X/n
    plt.stem(success_rates, pmf_vals, linefmt=color, markerfmt=color+'o', basefmt="k", label=f'n={n}')

# Formatting the plot
plt.xlabel('Success Rate (Y = X/n)')
plt.ylabel('Probability Mass Function')
plt.title('Success Rate Distribution for Different Campaign Sizes')
plt.legend()
plt.grid(True)
plt.show()
```

**Observations:**

1.For small

n (e.g., 10, 20 customers):

The success rate distribution is more spread out, meaning higher variability.
It is possible to observe a high success rate purely due to chance, even though the expected rate is 5%.

2.For larger

n (e.g., 50, 200 customers):

The success rate distribution becomes more concentrated around 5%.
The variance decreases, meaning the actual success rate is more predictable.

Implications for Large-Scale Marketing:

For small campaigns: There is a higher risk of observing misleading success rates due to small sample size.
For large campaigns: The results are more reliable and stable, making it easier to predict marketing performance.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given parameters
shape_params = [(5, 4), (10, 0.9), (7, 2), (1.5, 0.2)]

for alpha, theta in shape_params:
    print(f"\nFor Shape () = {alpha}, Scale () = {theta}:")

    # (a) Expected lifetime (Mean) and Variability (Variance)
    mean = alpha * theta
    variance = alpha * (theta**2)
    print(f"  Expected Lifetime (Mean) = {mean:} hours")
    print(f"  Variability (Variance) = {variance:}")

    # (b) Median Lifetime using inverse CDF
    median = stats.gamma.ppf(0.5, a=alpha, scale=theta)
    print(f"  Median Lifetime = {median:.2f} hours")

    # (c) Plot the probability density function (PDF)
    x = np.linspace(0.1, 50, 500)  # Range of lifetimes
    pdf = stats.gamma.pdf(x, a=alpha, scale=theta)

    plt.plot(x, pdf, label=f"={alpha}, ={theta}")

# Show the final plot
plt.xlabel("Battery Lifetime (hours)")
plt.ylabel("Probability Density")
plt.title("Gamma Distribution of Battery Lifetimes")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. A company has developed a new type of battery whose lifetime (in hours) follows a chi-squared
# distribution with 10 degrees of freedom. Using any programming language, answer the following questions:
from scipy.stats import chi2

# (a) Compute mean, standard deviation, median, Q1, and Q3
df = 10  # Degrees of freedom
mean = chi2.mean(df)
std_dev = chi2.std(df)
median = chi2.median(df)
q1 = chi2.ppf(0.25, df)
q3 = chi2.ppf(0.75, df)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
print(f"Median: {median}")
print(f"First Quartile (Q1): {q1}")
print(f"Third Quartile (Q3): {q3}")


# (b) Plot the battery lifetime distribution
x = np.linspace(0, 24, 500)
pdf = chi2.pdf(x, df)

plt.plot(x, pdf)
plt.title('Battery Lifetime Distribution')
plt.xlabel('Lifetime (hours)')
plt.ylabel('PDF')

plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label=f'Median = {median:.2f}')
plt.axvline(q1, color='b', linestyle='dashed', linewidth=1, label=f'Q1 = {q1:.2f}')
plt.axvline(q3, color='purple', linestyle='dashed', linewidth=1, label=f'Q3 = {q3:.2f}')

plt.legend()
plt.grid(True)
plt.show()

# (c) Moments using MGF (not directly using MGF but using existing functions for moments)
# The moment generating function is not used for computation. Rather, we use the built in functionality.
first_moment = chi2.moment(1, df)
second_moment = chi2.moment(2, df)
third_moment = chi2.moment(3, df)

print(f"First Moment: {first_moment}")
print(f"Second Moment: {second_moment}")
print(f"Third Moment: {third_moment}")
```

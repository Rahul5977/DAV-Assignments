```python
import numpy as np
import matplotlib.pyplot as plt

def count_islands(matrix):
    num_islands = 0
    largest_island_size = 0
    rows, cols = matrix.shape
    visited = np.zeros_like(matrix, dtype=bool)

    def iterative_dfs(row, col):
        stack = [(row, col)]
        island_size = 0

        while stack:
            r, c = stack.pop()
            if visited[r, c]:
                continue

            visited[r, c] = True
            island_size += 1

            for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
                new_r, new_c = r + dr, c + dc
                if (0 <= new_r < rows and 0 <= new_c < cols and
                        matrix[new_r, new_c] == 1 and not visited[new_r, new_c]):
                    stack.append((new_r, new_c))

        return island_size

    for i in range(rows):
        for j in range(cols):
            if matrix[i, j] == 1 and not visited[i, j]:
                num_islands += 1
                island_size = iterative_dfs(i, j)
                largest_island_size = max(largest_island_size, island_size)

    return num_islands, largest_island_size

probabilities = np.arange(0, 1.05, 0.05)
num_islands_list = []
largest_island_size_list = []

for p in probabilities:
    matrix = np.random.choice([0, 1], size=(1920, 1080), p=[1 - p, p])

    num_islands, largest_island_size = count_islands(matrix)

    num_islands_list.append(num_islands)
    largest_island_size_list.append(largest_island_size)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(probabilities, num_islands_list)
plt.xlabel("Probability (p)")
plt.ylabel("Number of Islands")
plt.title("Number of Islands vs. Probability")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(probabilities, largest_island_size_list)
plt.xlabel("Probability (p)")
plt.ylabel("Largest Island Size")
plt.title("Largest Island Size vs. Probability")
plt.grid(True)

plt.tight_layout()
plt.show()




```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading dataset
dataset = pd.read_csv('/content/DIstanceTimeDataset - StudentsHomeTownDistance.csv')

# Extracting and cleaning data
time = np.array(dataset['Time to Reach (hr)'].dropna())
distance = np.array(dataset['Distance (km)'].dropna())

# Initializing model parameters
slope = 35
intercept = 0
learning_rate = 0.0001  # Learning rate for stability
max_iters = 100

cost_values = []
iterations = []

# Gradient Descent Algorithm
iteration = 0
while iteration < max_iters:
    prev_slope = slope
    prev_intercept = intercept
    error_term = (prev_slope * time + prev_intercept - distance)  # Compute error term

    # Compute gradients
    grad_slope = np.mean(error_term * time)
    grad_intercept = np.mean(error_term)

    # Update parameters
    slope -= learning_rate * grad_slope
    intercept -= learning_rate * grad_intercept

    # Compute cost function
    updated_error = (slope * time + intercept - distance)
    cost = np.mean(updated_error ** 2) / 2
    cost_values.append(cost)
    iterations.append(iteration + 1)

    # Check for convergence
    prev_cost = np.mean(error_term ** 2) / 2
    if abs(prev_cost - cost) <= 0.01:
        print(f"Convergence reached at iteration {iteration} with cost {cost}")
        break

    iteration += 1

# Display final results
print(f"Optimized values: slope = {slope}, intercept = {intercept}")
print(f"Final cost = {cost}")

# Plot Cost vs Iterations
plt.figure(figsize=(8, 5))
plt.plot(iterations, cost_values, marker='o', linestyle='-')
plt.xlabel("Iteration")
plt.ylabel("Cost (J)")
plt.title("Cost Reduction Over Iterations")
plt.grid(True)
plt.show()
```
